## List

This is used to put a set of items in the list. Let's instantiate a list. 

```python
l = [1,2,3,4]
l1 = ["a", "b", "c", "d"]
l2 = ["d", 4, 3, 1.2, [1,2]]
print(l)
print(l1)
print(l2)
l.append(5)
print(l)
print(len(l))
```

Here's a few operation 

```python
print(l)
print(l[0])
print(l[-1])
print(l[:2])
print(l[1:])
print(l[1:3])
print(l[3:-1])
print(l[-1:])
print(l[5])
```

There is a few extra operations

```python
print(len(l))
l3 = list(range(1,10, 2))
print(l3)
print(sorted([5,2,1,5,3]))
sum(l)
print(l.count(5))
```

## Dictionary

Sometime you want a way to refer an item easily. That's where dictionary comes in. 

```python
d = {"a": 1, "b": 2, "c": 3}
print(d["b"])
```

You can put all type of things in a dict

```python
d1 = {"a":1, 3:1}
print(d1)
```

```python
d2 = {"a": 1, "b": [1,2,3], "c":{"a":1}}
```

```python
print(d2.keys())
print(d2.values())
```

```python
d3 = {1.2:1}
```

```python
people = {"name":"Ali", "birth_date":"1984", "organization":"codingshophouse"}
print(people["organization"])
print(people["name"])
```

## file object
I introduce this early because it is convenient. This is how a file work

```python
f = open("/tmp/simplefile.txt")
for i in f:
    print(i)
f.close()
```
